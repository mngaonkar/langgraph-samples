```python
from langgraph.prebuilt import ToolNode
from tools.visit_web_page_tool import VisitWebPageSyncTool
from langchain_core.messages import AIMessage
```

```python
tool_node = ToolNode([VisitWebPageSyncTool()])
```

```python
message_with_tool_call = AIMessage(
    content="",
    tool_calls=[
        {
            "name": "visit_webpage",
            "args": {
                "url": "https://www.thecloudcast.net",
                "clean_flag": True,
            },
            "type": "tool_call",
            "id": "tool_call_1",
        }
    ]
)
```

```python
result = tool_node.invoke({
    "messages": [message_with_tool_call],
})
```
